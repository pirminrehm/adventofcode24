```python
import numpy as np
import re
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
X_LEN = 103
Y_LEN = 101
ITERATIONS = 100
```

```python
points =  [tuple([int(x) for x in re.findall(r"-?\d+", l)]) for l in lines]
```

```python
counter = [0, 0, 0, 0]
for py, px, vy, vx in points:
    nx = (px + vx * ITERATIONS) % X_LEN
    ny = (py + vy * ITERATIONS) % Y_LEN
    if nx == X_LEN // 2  or ny == Y_LEN // 2 : continue
    ci = 0 if nx < X_LEN // 2 else 2
    ci += 0 if ny < Y_LEN // 2 else 1 
    counter[ci] += 1

part1 = 1
for c in counter:
    part1 *= c
part1
```

```python
for i in range(18, X_LEN * Y_LEN, 103):
    new_points = []
    for py, px, vy, vx in points:
        nx = (px + vx * i) % X_LEN
        ny = (py + vy * i) % Y_LEN
        new_points.append((nx, ny))
    grid = np.array([[" " for y in range(Y_LEN)] for x in range(X_LEN)], dtype=str)
    for p in new_points:
        grid[p] = "*"
    print(i)
    [print("".join(x)) for x in grid]
```
